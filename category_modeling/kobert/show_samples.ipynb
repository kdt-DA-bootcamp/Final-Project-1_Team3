```python
import pandas as pd

# CSV 파일 로드 (파일 경로를 실제 경로로 변경하세요)
df = pd.read_csv("C:/Users/hp/Desktop/Bootcamp/PROJECT_OTT_AARRR/clustered_postagged_kobert.csv")

# 각 클러스터별 대표 텍스트 출력
clusters = sorted(df["cluster"].unique())
for cluster in clusters:
    print(f"Cluster {cluster} 대표 샘플:")
    # 클러스터에 속하는 텍스트 중 상위 5개 출력
    sample_texts = df[df["cluster"] == cluster]["combined_text"].head(5)
    for text in sample_texts:
        print("-", text)
    print("-" * 50)
```
