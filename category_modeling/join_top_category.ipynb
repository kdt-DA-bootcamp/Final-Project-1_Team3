```python
import pandas as pd

# 파일 경로에 맞게 CSV 파일 읽기
df_categories = pd.read_csv('C:/Users/hp/Desktop/Bootcamp/PROJECT_OTT_AARRR/preprocessed_data_unique_updated.csv')
df_prep = pd.read_csv('C:/Users/hp/Desktop/Bootcamp/PROJECT_OTT_AARRR/pos_tagged_data.csv')

# pos_tagged_data에 기존에 있던 categoryID 열이 있다면 삭제
if 'categoryID' in df_prep.columns:
    df_prep = df_prep.drop(columns=['categoryID'])

# preprocessed_data_unique_updated.csv에서 videoID와 top_categoryID만 선택
df_map = df_categories[['videoID', 'categoryID']].copy()

# videoID 기준 inner join을 수행하여, preprocessed_data_unique_updated.csv에 존재하는 videoID에 대해서만 categoryID를 맵핑
df_merged = df_prep.merge(df_map, on='videoID', how='inner')

# 결과 CSV 파일로 저장
df_merged.to_csv('C:/Users/hp/Desktop/Bootcamp/PROJECT_OTT_AARRR/new_category_data.csv', index=False)
print("pos_tagged_data.csv에 새로운 categoryID가 맵핑된 CSV 파일이 저장되었습니다.")
```

```python
import pandas as pd

# 전처리된 데이터 파일 로드 (예시)
df_processed = pd.read_csv("preprocessed_data_comp_with_category_new.csv")

# videoID 별로 그룹화하여 각 videoID의 등장 횟수를 집계
videoid_counts = df_processed.groupby('videoID').size().reset_index(name='count')

# 한 videoID가 여러 번 등장한 경우 (즉, count > 1) 필터링
duplicated_videoids = videoid_counts[videoid_counts['count'] > 1]
print("중복된 videoID 및 등장 횟수:\n", duplicated_videoids)
```

```python
# 중복된 videoID를 담고 있는 Series (videoid_counts에서 count>1인 경우)
duplicated_videoids_list = duplicated_videoids['videoID'].tolist()

# df_processed에서 해당 videoID만 필터링
dup_mask = df_processed['videoID'].isin(duplicated_videoids_list)
df_duplicates = df_processed[dup_mask]

# 중복된 videoID들의 실제 행을 확인
df_duplicates.sort_values(by='videoID', inplace=True)
df_duplicates
```

```python
df_unique = df_processed.drop_duplicates(subset=['videoID'], keep='first')
print("중복 제거 전 행 수:", len(df_processed))
print("중복 제거 후 행 수:", len(df_unique))
df_unique.to_csv("preprocessed_data_unique.csv", index=False)
```
